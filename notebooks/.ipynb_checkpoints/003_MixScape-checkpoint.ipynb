## Loading Packages and Datasets

Loading all of the R packages that are needed to conduct perturbation quality control and the ECCITE-Seq merged replicate SCE object, target gene metadata, and differentially expressed genes results from the previous step.

```r
# Data Cleaning packages
library(tidyverse)
library(data.table)
library(hrbrthemes)

# Load packages.
library(Seurat)
library(scales)
library(ggplot2)
library(scRNAseq)
library(pheatmap)
library(patchwork)
library(colorspace)
library(DropletUtils)
library(BiocSingular)
library(SingleCellExperiment)

# Setup custom theme for plotting.
data_dir = '../data'
options(repr.plot.width=20, repr.plot.height=8)
custom_theme = theme(plot.title = element_text(size = 16, hjust = 0.5), legend.key.size = unit(0.7, "cm"), legend.text = element_text(size = 14))
```

```r
# Utility Function
extract_degs = function(results, matrix_genes){
    degs = list()
    
    for (gene in names(results)){
        gene_df = as.data.frame(results[[gene]])
        gene_df = gene_df[rownames(gene_df) %in% matrix_genes, ]
        gene_df = gene_df[!is.na(gene_df$padj), ]
        gene_df = gene_df[gene_df$padj < 0.05, ]
        degs[[gene]] = rownames(gene_df)
    }
    
    return (degs)
}
```

# Running MixScape Pipeline Manually

The following sectons of code is the individual steps of the Mixscape pipeline. On a merged set of cells from all the libraries.

```r
# load sce data
merged = readRDS(paste0(data_dir, '/SCE/Merged.Rds'))
merged = merged[, !duplicated(colnames(merged))]
merged = as.Seurat(merged)
```

```r
# load DEG data
res = readRDS(paste0(data_dir, '/DEGs/results/Merged.Rds'))
gene_metadata = read.csv(paste0(data_dir, '/Metadata/gene_metadata.csv'))
rownames(gene_metadata) = gene_metadata$Gene.Symbol

# extract merged degs
merged_degs = extract_degs(res, rownames(merged))

# get cell cycle genes
s.genes = cc.genes$s.genes
g2m.genes = cc.genes$g2m.genes
```

# Data Normalization

First I converted the merged SCE object into a Seruat Object as MixScape is a Seruat based method. After changing the object I renormalized the raw single cell gene expression values using a log scaling transformation (recommended by seruat) and found the highly variably expressed genes. Next I ran transformed values of highly expressed genes through PCA and UMAPs. The resulting visualization shows that the primary source immediately observable biological heterogenity comes from alternatives sources than gene perturbation (as observable in the second figure colored by target gene). Thereby demonstrates the necessity of perturbation quality control.

```r
# Prepare RNA assay for dimensionality reduction: Normalize data, find variable features and scale data.
DefaultAssay(object = merged) = "RNA"
merged = NormalizeData(object = merged) %>% FindVariableFeatures() %>% ScaleData()

# Run Principle Component Analysis (PCA) to reduce the dimensionality of the data.
merged = RunPCA(object = merged)

# Run Uniform Manifold Approximation and Projection (UMAP) to visualize clustering in 2-D.
merged = RunUMAP(object = merged, dims = 1:40)

# Generate plots to check if clustering is driven by biological replicate ID, cell cycle phase
# or target gene class.
p1 = DimPlot(merged, group.by = "batch", label = F, pt.size = 0.2, reduction = "umap") +
        scale_color_brewer(palette = "Dark2") + 
        ggtitle("Biological Replicate") +
        xlab("UMAP 1") + 
        ylab("UMAP 2") + 
        custom_theme

p2 = DimPlot(merged, group.by = "gene", label = F, pt.size = 0.2, reduction = "umap") + 
        ggtitle("Biological Replicate") +
        xlab("UMAP 1") + 
        ylab("UMAP 2") + 
        custom_theme
```

# Identifying Alternative Sources of Biological Heterogeneity

## Determining the Effects of the Cell Cycle via CellCycleScoring

This is the traditional method of removing the influence of the cell cycle from a given scRNA-Seq dataset. It follows a seruat tutorial were you first identify the phase of the cell cycle a given cell is in via scoring. Then taking the predicted cell cycle phase coeffiecients and regress the cell cycle out via the ScalingData function. As observed in the UMAP projection below the majority of the biological heterogeneity we observe is primarly due to the effect of the cell cycle. In certain aspect that really isn't all that surprising as our experiment is only made up of MCF7 cells the only major difference in gene expression on a large scale is most likely the cell cycle.

```r
merged = CellCycleScoring(merged, s.features=s.genes, g2m.features=g2m.genes, set.ident=TRUE)
```

```r
# using the default list of cell cycle markers that seruat provides
RidgePlot(merged, features = c("PCNA", "TOP2A", "MCM6", "MKI67"), ncol = 2)

p_cell_cycle = DimPlot(merged, group.by = "Phase", label = F, pt.size = 0.2, reduction = "umap") + 
                ggtitle("Distribution of Cell Cycle Phase") +
                xlab("UMAP 1") + 
                ylab("UMAP 2") + 
                custom_theme

p_cell_cycle
```

# Calculating Perturbation Signatures

In order, to remove the effects of alternative sources of biological heterogenity Papalexi et al. 2020 proposed and developed perturbation signatures to remove all cell cycle and cell stress effets from biological heterogenity. In order to calculate these signatures take the average expression profile of the K (K=20 in this case) closet non-targeting gRNA contain cells for a given cells containing a gRNA. Then you subtract the K ctrl averaged expression profile from a theoritically perturbed cell the resulting expression profile is called the perturbation signature.

Fundamentally, this is a normalization technique that aims to remove the effects from cell stress induced by viral infection, cell cycle, and batch effects. While, i have questions about the optimal K ctrl cells to use this normalization technique is adequte at removing stochastic noise. 

As you can see perturbation signatures removed all of the alternative sources of biological heterogenity that was previously obvserved above. This only highlights and reinforces the importance of remove cell cycle effects from the scRNA-Seq CRISPR data. However, we not completely done yet.

```r
perturb_qc_fp = paste0(data_dir, '/SCE/Merged_mixscape.Rds')

if (!file.exists(perturb_qc_fp)) {
    # Calculate local perturbation signature.
    merged = CalcPerturbSig(object = merged, assay = "RNA", slot = "data", gd.class = "gene", nt.cell.class = "Control", 
                           reduction = "pca", ndims = 40, num.neighbors = 20, split.by = "batch", new.assay.name = "PRTB")

    # Prepare PRTB assay for dimensionality reduction: Normalize data, find variable features and center data.
    DefaultAssay(object = merged) = "PRTB"

    # Use variable features from RNA assay.
    VariableFeatures(object = merged) = VariableFeatures(object = merged[["RNA"]])
    merged = ScaleData(merged, do.scale = F, do.center = T)

    # Run PCA to reduce the dimensionality of the data and UMAP to visualize clustering in 2-D
    merged = RunPCA(object = merged, reduction.key = "prtbpca", reduction.name = "prtbpca")
    merged = RunUMAP(object = merged, dims = 1:40, reduction = "prtbpca", reduction.key = "prtbumap", reduction.name = "prtbumap")

    saveRDS(merged, perturb_qc_fp)
} else {
    # load the mixscape qc sce object
    merged <- readRDS(perturb_qc_fp)
}
```

```r
p1 = DimPlot(merged, group.by = "batch", reduction = "prtbumap", pt.size = 0.2) + 
        scale_color_brewer(palette = "Dark2") + 
        ggtitle("Visualizing Biological Replicate in Perturbation Signatures") + 
        ylab("UMAP 2") + 
        xlab("UMAP 1") + 
        custom_theme

p2 = DimPlot(merged, group.by = "Phase", reduction = "prtbumap", pt.size = 0.2) + 
        ggtitle("Visualizing Cell Cycles in Perturbation Signatures") + 
        ylab("UMAP 2") + 
        xlab("UMAP 1") + 
        custom_theme

p1 | p2
```

# Running Mixscape on Perturbation Signatures

Now that we have removed all of alternative sources of biological heterogenity we can now attemtp to differentate between perturbed and non-perturbed cells. In order, to do this we're going to use Papelexi et al. 2020 MixScape function. Mixscape is a Mixture Modeling method for identify and separate transcriptional pertured and non-perturbed cells. I've currently modified it's code so I used the DEGs from DESeq2 the original version of MixScape conducted Differentially Expression Analysis using Welch t-test. I found the t-test to be in sensitive and ultimately provide a subpar performance.

```r
# extract the list of differentially expressed genes
perturb_qc_fp = paste0(data_dir, '/SCE/Merged_mixscape.Rds')

if (!'mixscape_class.global' %in% colnames(merged[[]])) {
#   Run mixscape to classify cells based on their perturbation status.
  merged = RunMixscape(object = merged, assay = "PRTB", slot = "scale.data", labels = "gene", 
                        de.gene.list = merged_degs, provided.de.list = TRUE, nt.class.name = "Control",
                        min.de.genes = 20, iter.num = 10, de.assay = "RNA", verbose = F)
  
  saveRDS(merged, perturb_qc_fp)
} else {
  # load the mixscape qc sce object
  merged = readRDS(perturb_qc_fp)
}
```

# Visualizing Perturbed Cells

Now that MixScape has identified Non-Perturbed (NP) and Perturbed (KO) cells we can filter out the NP cells. Once we've removed all of the NP cells we can attempt to visualize genetic perturbations using the MixscapeLDA function. The function uses a linear discriminator analysis to separate genetic perturbations (similar to a PCA) a PCA will also work. Finally, after conducting LDA we use UMAP to compresse the dimensionality into a 2 dimesions and visualize.

```r
visualize_perturbations <- function(sce, ndims=9) {
    # Remove non-perturbed cells and run LDA to reduce the dimensionality of the data.
    Idents(sce) = "mixscape_class.global"
    gene_sub = subset(sce, idents = c("KO", "Control"))

    # Running Linear Discriminate Analysis to identify effects of perturbing target genes 
    gene_sub = MixscapeLDA(object = gene_sub, assay = "RNA", pc.assay = "PRTB", labels = "gene", 
                           nt.label = "Control", npcs = 10, logfc.threshold = 0.25, verbose = F)

    # Use LDA results to run UMAP and visualize cells on 2-D.
    gene_sub = RunUMAP(gene_sub, dims = 1:ndims, reduction = "lda", reduction.key = "ldaumap", reduction.name = "ldaumap")

    # Visualize UMAP clustering results.
    Idents(gene_sub) = "mixscape_class"
    gene_sub$mixscape_class = as.factor(gene_sub$mixscape_class)
    
    return (gene_sub)
}
```

```r
options(repr.plot.width=8, repr.plot.height=8)
top_11_genes <- c('Control', 'TFAP2C', 'MYC', 'GATA3', 'MTOR', 'SMC1A', 'ESR1', 'FOXA1', 'NCOA3', 'CREBBP', 'MED1', 'CCND1')
cells_to_keep <- merged[[]] %>% filter(gene %in% top_11_genes) %>% rownames

gene_sub = visualize_perturbations(merged[, cells_to_keep])
perturb_umap_colour_vals <- hue_pal()(14)
perturb_umap_colour_vals[3] <- "Grey"


perturb_umap = DimPlot(gene_sub, reduction = "ldaumap", group.by = 'gene',
                       label = T, repel = T, label.size = 5) +
                        scale_colour_manual(values = perturb_umap_colour_vals) +
                        xlim(c(-5, 5)) +
                        ylim(c(-5, 5)) + 
                        coord_fixed() +
                        ggtitle("Visualizing Perturbation Effect")

p_cell_cycle = DimPlot(merged, group.by = "Phase", label = F, pt.size = 0.2, reduction = "umap") + 
                ggtitle("Distribution of Cell Cycle Phase") +
                xlab("UMAP 1") + 
                ylab("UMAP 2") + 
                coord_fixed() +
                custom_theme

p2 = DimPlot(merged, group.by = "gene", label = F, pt.size = 0.2, reduction = "umap") + 
        ggtitle("Biological Replicate") +
        xlab("UMAP 1") + 
        ylab("UMAP 2") + 
        coord_fixed() +
        custom_theme
```
